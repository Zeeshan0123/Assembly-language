# **Data Prepration**

```python
# We now use the Llama-3.1 format for conversation style finetunes
from unsloth.chat_templates import get_chat_template

tokenizer = get_chat_template(
    tokenizer,
    chat_template = "llama-3.1",
)
```

```python
dataset = [
  {
    "question": "What is the model number of the refrigerator part being described?",
    "answer": "A2RXNMFWW00, A2RXNMFWB00, A2RXNMFWS00"
  },
  {
    "question": "Are the parts available in different colors?",
    "answer": "Yes"
  },
  {
    "question": "What are the three colors in which these cabinet parts are available?",
    "answer": "White, Black, Stainless Steel"
  },
  {
    "question": "Is this part printed with a lithograph process?",
    "answer": "Yes"
  }
]
```

```python
# Step 1: Convert to desired format
converted_dataset = []
for item in dataset:
    conversation = [
        {"role": "user", "content": item["question"]},
        {"role": "assistant", "content": item["answer"]}
    ]
    converted_dataset.append(conversation)

# Wrapping converted_dataset in a dictionary to mimic the structure you need
formatted_examples = {"conversations": converted_dataset}
```

```python
from datasets import Dataset

# Step 2: Create a list of dictionaries with a 'conversation' key
dataset_for_hf = [{"conversation": conv} for conv in converted_dataset]

# Step 3: Create the Hugging Face Dataset
dataset = Dataset.from_list(dataset_for_hf)
```

```python
dataset
```

```python
# Step 4: Define formatting function
def formatting_prompts_func(examples):
    convos = examples["conversation"]
    texts = [tokenizer.apply_chat_template(convo, tokenize=False, add_generation_prompt=False) for convo in convos]
    return {"text": texts}

# Step 5: Apply map with batched=True
dataset = dataset.map(formatting_prompts_func, batched=True)

# View the output
print(dataset)
```

```python
dataset[2]["text"]
```

```python
from trl import SFTTrainer
from transformers import TrainingArguments, DataCollatorForSeq2Seq
from unsloth import is_bfloat16_supported
```

```python
# The DataCollatorForSeq2Seq is designed for sequence-to-sequence tasks
# where you have input sequences and target sequences (like machine translation).
# It prepares your data for this type of training by padding and creating attention masks
# for both the input and target sequences. However, in your case, using SFTTrainer
# for instruction fine-tuning usually only involves a single text field representing
# the instructions and responses.

# By using DataCollatorForSeq2Seq, you were likely providing data in a format
# that the SFTTrainer and your underlying model were not expecting.
# Removing it allows the SFTTrainer to handle the data preparation in a way
# that is more suitable for instruction fine-tuning, where you only need to process
# a single text field.

trainer = SFTTrainer(
    model = model,
    tokenizer = tokenizer,
    train_dataset = dataset,
    dataset_text_field = "text",
    max_seq_length = max_seq_length,
    # data_collator = DataCollatorForSeq2Seq(tokenizer = tokenizer),
    dataset_num_proc = 2,
    packing = False, # Can make training 5x faster for short sequences.
    args = TrainingArguments(
        per_device_train_batch_size = 2,
        gradient_accumulation_steps = 4,
        warmup_steps = 5,
        # num_train_epochs = 1, # Set this for 1 full training run.
        max_steps = 10,
        learning_rate = 2e-4,
        fp16 = not is_bfloat16_supported(),
        bf16 = is_bfloat16_supported(),
        logging_steps = 1,
        optim = "adamw_8bit",
        weight_decay = 0.01,
        lr_scheduler_type = "linear",
        seed = 3407,
        output_dir = "outputs",
        report_to = "none", # Use this for WandB etc
    ),
)
```

```python
#@title Show current memory stats
gpu_stats = torch.cuda.get_device_properties(0)
start_gpu_memory = round(torch.cuda.max_memory_reserved() / 1024 / 1024 / 1024, 3)
max_memory = round(gpu_stats.total_memory / 1024 / 1024 / 1024, 3)
print(f"GPU = {gpu_stats.name}. Max memory = {max_memory} GB.")
print(f"{start_gpu_memory} GB of memory reserved.")
```

```python
trainer_stats = trainer.train()
```
